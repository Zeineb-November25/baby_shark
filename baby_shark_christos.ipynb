```python
# I choose to cluster the values in series "Activity" by

# surfing
# swimming
# fishing (includes spearfhising)
# diving/snorkeling
# boating
# unknown

# All values will be tranformed to the closest relation with these words. 
# Disclaimer: All activities on something bigger than a board will be clustered under boating.
# All water activities like standing, playing, trading water, bathing, etc will go under swimming
# All diving activities will be merged regardless of the nature of diving.
# Snorkeling will be examined differently than swimming, and together with diving, because snorkeling by definition takes place in ecosystems where a lot of fish exist, and therefore, a higher probability of sharks.
```

```python
# now I proceed with clustering
# I start with an activity that I want it to be transformed into a category that is later in the sequence of ifs, but contains a word that will be used for surfing categorization, hence, I put that one first.

def map_activity_to_cluster(x: str) -> str:
    x = str(x).lower().strip()
    
    if "fell overboard" in x:
        return "boating"
    if any(word in x for word in ["surf","paddle", "boarding"]):
        return "surfing"
    if any(word in x for word in ["swim", "wading", "bathing", "treading", "standing", "floating", "playing"]):
        return "swimming"
    if "fish" in x:
        return "fishing"
    if any(word in x for word in ["dive", "snorkel", "diving"]):
        return "diving/snorkeling"
    if any(word in x for word in ["boat", "kayak", "canoe", "overboard", "rowing", "sailing"]):
        return "boating"
    if "unknown" in x:
        return "unknown"
    return "other"
```

```python
# applying the function to the series

baby_shark_christos_df["Activity_cluster"] = baby_shark_christos_df["Activity"].apply(map_activity_to_cluster)
```
